```python
import re
def remove_punctuation(text):
    return re.sub(r'[^\w\s]', '', text)
data['text'] = data['text'].apply(remove_punctuation)
data
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
data["test"] = data.text.apply(word_tokenize)
data
```

```python
empty_rows = data['text'] == '[EMPTY] '
if empty_rows.any(  ):
     data.drop(index=empty_rows[empty_rows].index,inplace=True)
data
```

```python
from nltk.corpus import stopwords
stop_words=stopwords.words('english')
print(stop_words)
```

```python
stop_words = set(stopwords.words('english'))

def test(word_tokens):
    filtered_sentence = [w for w in word_tokens if not w.lower() in stop_words]
    return filtered_sentence

data["test2"] = data.test.apply(test)
data
```

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
def stem_list_of_strings(words_list):
    return [ps.stem(word) for word in words_list]

data['stemmed_test2'] = data['test2'].apply(lambda x: stem_list_of_strings(x))
data
```

```python
from nltk.stem import WordNetLemmatizer

wnl = WordNetLemmatizer()
def stem_list_of_strings(words_list):
    return [wnl.lemmatize(word) for word in words_list]


data['lemmatize3'] = data['test2'].apply(lambda x: stem_list_of_strings(x))
data
def convert_to_string(listReview):
    return '  '.join(listReview)

data['final_processed']=data['lemmatize3'].apply(convert_to_string)

data
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
import pandas as pd
atc_keywords = {
    "Radar Contact": ["radar contact", "identified", "squawk ident", "paint"],
    "Climb Instruction": ["climb", "level", "flight level", "maintain", "positive rate", "cleared to"],
    "Speed Adjustment": ["speed", "knots", "mach", "reduce", "increase", "maintain"],
    "General Information": ["advise intentions", "report position", "report level", "information", "be advised", "expect"],
    "Descend Instruction": ["descend", "level", "flight level", "maintain", "cleared to"],
    "Frequency Change": ["contact", "monitor", "squawk", "listen", "frequency", "change"],
    "Holding": ["hold", "expect further clearance", "pattern", "entry", "instructions"],
    "Approach": ["cleared for approach", "cleared ILS", "cleared visual approach", "cleared runway"],
    "Departure": ["cleared for takeoff", "taxi", "hold short", "departure frequency", "cleared to climb"],
    "Route Clearance": ["cleared to", "via", "direct", "maintain", "flight level", "track", "heading"],
    "Weather Information": ["windshear", "turbulence", "icing", "visibility", "clouds"],
    "Emergency": ["declare emergency", "squawk 7700", "maintain heading and altitude", "divert"]
}

def classify_text(text, atc_keywords):

    text = text.lower()

    for instruction_type, keywords in atc_keywords.items():
        for keyword in keywords:
            if keyword in text.split():
                return instruction_type

    return "General Information"

data['category'] = data['final_processed'].apply(classify_text, args=(atc_keywords,))

category_distribution = data['category'].value_counts()
#rawda
print(data.head())
print(category_distribution)
data.to_csv("data_after_pattern1.csv", index=False)
```
